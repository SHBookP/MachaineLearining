````python
x = [i*0.2 for i in range(40)]
y = [i**3-13*i**2+47*i-35 for i in x]

plt.plot(x,y, label="$x^3-13x^2+47x-35$",color="red",linewidth=2)
plt.xlabel("x", fontsize=26)
plt.ylabel("y", fontsize=26)
plt.title("text标注例子", fontsize=26, color='green')
plt.text(1, -20, "$x^3-13x^2+47x-35$", fontsize=26, color='purple')
plt.show()
````

````python
x = [i*0.2 for i in range(40)]
y = [i**3-13*i**2+47*i-35 for i in x]

plt.plot(x,y, label="$x^3-13x^2+47x-35$",color="red",linewidth=2)
plt.xlabel("x", fontsize=26)
plt.ylabel("y", fontsize=26)
plt.title("annotate标注例子", fontsize=26, color='green')
# shrink为小于1的数字，表示收缩箭头的幅度（离开文字以及指定端点有多远）。
plt.annotate('极小值', xy=(6.1, -5), xytext=(5, -25),
            arrowprops=dict(facecolor='black', shrink=0.1), fontsize=26)
plt.show()
````

## imshow、等高线
imshow 坐标默认是从上到下，从左到右。使用参数 origin='lower' 可以把坐标原点移到左下角。

````python
import math
z = []
for i in range(240):
    x = -3. + i * 0.025
    ly = []
    for j in range(240):
        y = -3. + j * 0.025
        v = math.exp(-1 * x**2 - y**2) - math.exp(-(x-1)**2 -(y-1)**2)
        ly.append(v)
    z.append(ly)
````

````python
im = plt.imshow(z, origin='lower')
plt.colorbar(im)
````

````python
levels = [-0.8, -0.6, -0.4, -0.2, 0, 0.2, 0.4, 0.6, 0.8]
plt.contour(z, levels)
````

## 随机游走

````python
import random
import matplotlib.pyplot as plt

N = 10000
# 产生-1或1的列表
move = [random.randrange(-1,2,2) for i in range(N)]

# 累加
s = []
cumsum = 0
for item in move:
    cumsum += item
    s.append(cumsum)
    
plt.plot(s)
plt.show()
````

## 画出地震深度、震级的分布图
earthquake_data.csv 是2013.1-6月的数据。格式为： 日期,时间,纬度(°),经度(°),深度(km),震级类型,震级值,事件类型,参考地名。 例子：

2013-06-23,23:26:20.8,33.72,119.90,12,ML,1.7,eq,江苏阜宁
2013-06-23,23:25:03.7,44.71,82.14,6,ML,1.7,eq,新疆博乐
2013-06-23,23:15:14.3,30.90,103.43,11,ML,1.2,eq,四川汶川
2013-06-23,22:55:33.4,30.38,102.96,16,ML,2.5,eq,四川芦山
2013-06-23,22:39:44.8,24.70,99.22,5,ML,1.6,eq,云南施甸
2013-06-23,22:31:30.3,24.70,99.21,5,ML,1.4,eq,云南施甸

取出其中的深度、震级信息，画出其分布图。

````python
import matplotlib.pyplot as plt

with open('earthquake_data.csv', 'r') as f:  # 自动调用close()
    lines = f.readlines()[1:]
    
extract = []
for line in lines:
    line = line.strip('\n').split(',')
    extract.append([float(line[4]), float(line[6]), line[7]])

deep  = [i[0] for i in extract]
amplitude = [i[1] for i in extract]

plt.subplot(211)
plt.hist(deep)
plt.subplot(212)
plt.hist(amplitude)
plt.show()
````

地震深度分布很不均匀，两种处理。
1. 观察更小的范围```[0,100)```
2. 取对数后观察

````python
max(deep)
````

````python
min(deep)
````

````python
deep2 = [i for i in deep if i < 100]
plt.hist(deep2)
plt.show()
````

````python
import math
deep3 = [math.log(i+1) for i in deep]
plt.hist(deep3)
plt.show()
````
