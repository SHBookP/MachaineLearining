# 第一题，2-D随机游走

```python
import matplotlib.pyplot as plt
import numpy as np

x,y=[0],[0]

for n in range(1000):
    if np.random.choice([-1,1])==-1:
        x.append(x[n]+np.random.choice([-1,1]))
        y.append(y[n])
    else:
        y.append(y[n]+np.random.choice([-1,1]))
        x.append(x[n])
plt.plot(x,y)
plt.show()
```

# 第二题，弹性球运动轨迹

```python
import matplotlib.pyplot as plt
import numpy as np

h0,vx0,g=2,1,-9.8
v=np.array([vx0,0])
r=np.array([[0,h0]])
a=np.array([0,g])

t=0
dt=.0001

total_t=np.sqrt(2*2./9.8)*(5+2*np.sqrt(2))

while(t<=total_t):
    if(r[-1,1]<0):
        v[0],v[1]=v[0]/np.sqrt(2),-v[1]/np.sqrt(2)
        r[-1,1]=0
    r=np.append(r,[r[-1]+v*dt+.5*a*dt**2],axis=0)
    v=v+a*dt
    t+=dt

plt.plot(r[:,0],r[:,1])
plt.show()
```
