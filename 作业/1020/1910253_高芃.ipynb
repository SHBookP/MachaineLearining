```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def Turn(x,d,deg):
    ''' 输入：x：之前位移矢量；d：转动距离；deg：转动角度'''
    x_deg=np.arctan2(x[1],x[0])
    Deg=x_deg+deg/180*np.pi
    return d*np.array([np.cos(Deg),np.sin(Deg)]) 

r=np.array([[0,0]])
r=np.append(r,[[1,0]],axis=0)

for n in range(2,360+1):
    x=r[-1]-r[-2]
    r=np.append(r,[r[-1]+Turn(x,n,59)],axis=0)
    
plt.plot(r[:,0],r[:,1])
plt.xlabel("x/ mm")
plt.ylabel("y/ mm")
plt.show()
```
