```python
import matplotlib.pyplot as plt
import numpy as np

a=1.4
b=.3

xy=np.array([[.1,.2]])
for n in range(8191):
    xy=np.append(xy,[[1-a*xy[-1,0]**2+xy[-1,1],b*xy[-1,0]]],axis=0)

plt.plot(xy[:,0],xy[:,1],'.')
plt.title("Hénon")
plt.show()

xmin,xmax=.4,1
ymin,ymax=-.3,.3
for t in range(1,6):
    time=t**2.2
    plt.plot(xy[:,0],xy[:,1],'.')
    plt.title("Hénon"+str(t))
    plt.xlim([(xmin+xmax)/2-(xmax-xmin)/(2*time),(xmin+xmax)/2+(xmax-xmin)/(2*time)])
    plt.ylim([(ymin+ymax)/2-(ymax-ymin)/(2*time),(ymin+ymax)/2+(ymax-ymin)/(2*time)])
    plt.show()
```
