## 第一题，最后一行输出结果

```python
import pandas as pd
data=pd.read_csv('空气质量.csv')
tianjin=data.loc[data['city']=="天津"]
tianjin
```

## 第二题，最后一行输出结果

```python
pm25=tianjin.sort_values(by=['PM2.5'],ascending=False).head(200)
pm25
```

## 第三题，print出结果

```python
date=pm25['time'].str.split('-',expand=True)
rlt=date[1].value_counts()
print('是第'+rlt.index[0]+'月出现最多，为'+repr(rlt[0])+'次')
```
